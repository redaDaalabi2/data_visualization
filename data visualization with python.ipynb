```python
x = [6,2,9]
y = [1,7,8]
plt.title("graph")
plt.plot(x)
plt.plot(y)
plt.xlabel(x)
plt.ylabel(y)
plt.legend(['this is x', 'this is y'])
plt.show()
```

```python
sample_countries = pd.read_csv('countries.csv')
```

```python
sample_countries
```

```python
Afghanistan = sample_countries[sample_countries.country == 'Afghanistan']
```

```python
Afghanistan
```

```python
Zimbabwe = sample_countries[sample_countries.country == 'Zimbabwe']
```

```python
Zimbabwe
```

```python
plt.title("compare population of two country")
plt.plot(Afghanistan.year, Afghanistan.population / 10**6, 'o')
plt.plot(Zimbabwe.year, Zimbabwe.population / 10**6, 'o')
plt.xlabel('year')
plt.ylabel('population')
plt.legend(['this is afghanistan', 'this is zimbabwe'])
plt.show()
```
